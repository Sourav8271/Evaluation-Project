# Here we analyse top 10 cusines in each country through the basis of votes

```python
# Function to get top 10 cuisines for each country
def get_top_cuisines(df):
    top_cuisines = pd.DataFrame()
    for country in df['Country'].unique():
        country_data = df[df['Country'] == country]
        top_country_cuisines = country_data.groupby('Cuisines')['Votes'].sum().reset_index()
        top_country_cuisines = top_country_cuisines.sort_values(by='Votes', ascending=False).head(10)
        top_country_cuisines['Country'] = country
        top_cuisines = pd.concat([top_cuisines, top_country_cuisines], ignore_index=True)
    return top_cuisines

# Get the top cuisines
top_cuisines = get_top_cuisines(merged_df)

# Function to plot top 10 cuisines for each country
def plot_top_cuisines(top_cuisines):
    countries = top_cuisines['Country'].unique()
    num_countries = len(countries)

    # Create a figure with subplots for each country
    fig, axes = plt.subplots(num_countries, 1, figsize=(12, num_countries * 5), constrained_layout=True)

    # If there's only one country, axes is not a list, so we need to handle that
    if num_countries == 1:
        axes = [axes]

    # Loop through each country and create a bar plot for its top cuisines
    for i, country in enumerate(countries):
        country_data = top_cuisines[top_cuisines['Country'] == country]
        sns.barplot(x='Votes', y='Cuisines', data=country_data, ax=axes[i], palette='viridis')
        axes[i].set_title(f'Top 10 Cuisines in {country}')
        axes[i].set_xlabel('Votes')
        axes[i].set_ylabel('Cuisines')

    plt.show()

# Plot the top cuisines
plot_top_cuisines(top_cuisines)
```

# Using IQR method to remove outliers

```python
# Calculate IQR
columns_of_interest = ['Price range','Aggregate rating','Votes']

Q1 = merged_df[columns_of_interest].quantile(0.25)
Q3 = merged_df[columns_of_interest].quantile(0.75)
IQR = Q3 - Q1

# Determine outliers based on IQR
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = merged_df[((merged_df[columns_of_interest] < lower_bound) | (merged_df[columns_of_interest] > upper_bound)).any(axis=1)]

# Remove outliers
df1  = merged_df[~((merged_df[columns_of_interest] < lower_bound) | (merged_df[columns_of_interest] > upper_bound)).any(axis=1)]

# Print cleaned DataFrame
print("DataFrame after removing outliers:")
print(df1)

# Optional: Print removed outliers
print("\nOutliers removed:")
print(outliers)
```

# Find the one of best model

```python
# importing libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.linear_model import Lasso,Ridge
from sklearn.svm import SVR
from sklearn.ensemble import ExtraTreesRegressor
```

# Find the best random state and accuracy

```python
# Finding the best random state
maxAccu = 0
maxRS = 0
for i in range(1, 200):
    # Splitting the data into train and test sets using current random state (i)
    x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.30, random_state=i)
    
    # Initializing and training the Gradient Boosting Regressor
    RFR = RandomForestRegressor()
    RFR.fit(x_train, y_train)
    pred = RFR.predict(x_test)
    acc=r2_score(y_test,pred)
    if acc>maxAccu:
        maxAccu = acc
        maxRS = i
print("Maximum r2 score is ",maxAccu,"on Random_State",maxRS)
```

Here we can see that almost 100 % accuracy
